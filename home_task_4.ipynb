#### 4.3 Работа с датасетом "house price"
4.1 Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
    - Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
price_corr = round(house_df.corr(), 2)['price']
price_corr
```

```python
total_price_corr = price_corr[price_corr > 0.4]
total_price_corr
```

```python
pd.plotting.scatter_matrix(house_df[total_price_corr.index]);
```

```python
house_df.plot(kind='scatter', x='bathrooms', y='price', title='Диаграмма рассеивания зависимости цены от количеств ванных комнат');
```

```python
house_df.groupby('bathrooms')['price'].mean().plot(kind='line', 
                                                   title='Средняя цена домов в зависимости от количества ванных', 
                                                   xlabel='Количестко ванных комнат', 
                                                   ylabel='Цена в млн.');
```

```python
house_df.groupby('bathrooms')['price'].median().plot(kind='line', 
                                                   title='Медианная цена домов в зависимости от количества ванных', 
                                                   xlabel='Количестко ванных комнат', 
                                                   ylabel='Цена в млн.');
```

Очевидная зависимость.

Просадка в цене на недвижимость с 7 ванными комнатами скорее всего является выбросом судя по диаграмме рассеивания.

Разница между графиками средней и медианной цены незначительны и не влияют на общий вывод.

```python
house_df.plot(kind='scatter', x='sqft_living', y='price', alpha = 0.2, 
              title='Диаграмма рассеивания зависимости цены от жил площади',
              xlim=(0,4000), # Убираем выбросы по жил площади
              ylim=(0,1500000), # Убираем выбросы по цене
              xlabel='Жил площадь',
              ylabel='Цена');
```

Диаграмма рассеивания зависимости цены от жил площади наглядно показывает прямую зависимость между признаками.

```python
house_df.plot(kind='scatter', x='grade', y='price',
              title='Диаграмма рассеивания зависимости цены от оценки дома',
              xlabel='Оценка',
              ylabel='Цена в млн.');
```

```python
house_df.groupby('grade')['price'].mean().plot(kind='line', style='-',
                                                   title='Средняя цена домов в зависимости от оценки', 
                                                   xlabel='Оценка дома', 
                                                   ylabel='Цена в млн.');
```

```python
house_df.groupby('grade')['price'].median().plot(kind='line', style='-',
                                                   title='Медианная цена домов в зависимости от оценки', 
                                                   xlabel='Оценка дома', 
                                                   ylabel='Цена в млн.');
```

Графики медианной и средней цены в зависимости от оценки дома показывает очень сильную зависимость.

В следствии чего можно рассматривать показатель оценки как основной при формировании цены.
